```python
import pandas as pd
```

```python
# Getting the postal codes corresponding to Borough
postalcodes = 'https://raw.githubusercontent.com/muzabu51/Coursera_Capstone/master/Canada_pcodes.csv'
df_postalcodes = pd.read_csv(postalcodes, index_col=0)
df_postalcodes
```

```python
# Getting Geodata from file since geocoder package doesn't work as expected
Geodata = 'http://cocl.us/Geospatial_data'
df_geo = pd.read_csv(Geodata)
df_geo
```

```python
df_merged = pd.merge(df_postalcodes, df_geo, on='Postal Code')
df_merged
```

```python
df_merged = df_merged.groupby(df_merged['Postal Code']).first()
df_merged
```

```python
df_merged.to_csv(r'Canada_post_lat_long.csv')
```
